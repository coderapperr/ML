```python
feature_data = data.iloc[:,:23]
feature_data = feature_data.drop(['Div','Date'],axis=1)
```

```python
feature_data.head()
```

```python
#Team, Home Goals Scored, Away Goals Scored, Attack Strength, Home Goals Conceded, Away Goals Conceded, Defensive Strength
table = pd.DataFrame(columns=('Team','HGS','AGS','HAS','AAS','HGC','AGC','HDS','ADS',))
table = table[:-10]
```

```python
feature_data = feature_data[:-10]
feature_data.shape[0]
```

```python
feature_data_home = feature_data.groupby('HomeTeam')
feature_data_away = feature_data.groupby('AwayTeam')
```

```python
#list(feature_data_home)[0]
hometeam=list(feature_data_home.HomeTeam)
team=[]
for i in hometeam:
    team.append(i[0])
table.Team = team    
```

```python
table.HGS = feature_data_home.FTHG.sum().values
table.HGC = feature_data_home.FTAG.sum().values
table.AGS = feature_data_away.FTAG.sum().values
table.AGC = feature_data_away.FTHG.sum().values
```

```python
num_games = feature_data.shape[0]/20
```

```python
avg_home_scored = feature_data.FTHG.sum()*1.0 / feature_data.shape[0]
avg_away_scored = feature_data.FTAG.sum()*1.0 / feature_data.shape[0]
avg_home_conceded = avg_away_scored
avg_away_conceded = avg_home_scored
```

```python
table.HAS = (table.HGS / num_games) / avg_home_scored
table.AAS = (table.AGS / num_games) / avg_away_scored
table.HDS = (table.HGC / num_games) / avg_home_conceded
table.ADS = (table.AGC / num_games) / avg_away_conceded
table.head()
```

```python
table[table.Team == "Arsenal"]
```

```python
feature_data.columns
```

```python
feature_table = feature_data.drop(['FTHG','FTAG','HTHG','HTAG','Referee',],axis=1)
f_HAS = []
f_HDS = []
f_AAS = []
f_ADS = []
for index,row in feature_table.iterrows():
    f_HAS.append(table[table['Team'] == row['HomeTeam']]['HAS'].values[0])
    f_HDS.append(table[table['Team'] == row['HomeTeam']]['HDS'].values[0])
    f_AAS.append(table[table['Team'] == row['AwayTeam']]['AAS'].values[0])
    f_ADS.append(table[table['Team'] == row['AwayTeam']]['ADS'].values[0])
    
feature_table['HAS'] = f_HAS
feature_table['HDS'] = f_HDS
feature_table['AAS'] = f_AAS
feature_table['ADS'] = f_ADS
```

```python
feature_table.head(20)
```

```python
def transformFTResult(row):
    '''Converts results (H,A or D) into numeric values'''
    if(row.FTR == 'H'):
        return 1
    elif(row.FTR == 'A'):
        return -1
    else:
        return 0
    
def transformHTResult(row):
    '''Converts results (H,A or D) into numeric values'''
    if(row.HTR == 'H'):
        return 1
    elif(row.HTR == 'A'):
        return -1
    else:
        return 0    
```

```python
feature_table["HTResult"] = feature_table.apply(lambda row: transformHTResult(row),axis=1)
feature_table["FTResult"] = feature_table.apply(lambda row: transformFTResult(row),axis=1)
```

```python
feature_table.head(20)
```

```python
X = feature_table.drop(['HomeTeam','AwayTeam','FTR','HTR','FTResult'],axis=1)
y = feature_table['FTResult']
```

```python
X.head()
#y.head()
```

PREDICTION:


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print(accuracies.mean())
print(accuracies.std())
```
